```python
# Definir modelos y grids de hiperparámetros
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier

models_dict = {
    'LR': (LogisticRegression(max_iter=1000), {'C': [0.1, 1, 10]}),
    'NB': (GaussianNB(), {}),
    'KNN': (KNeighborsClassifier(), {'n_neighbors': [3, 5, 7]}),
    'SVM': (SVC(probability=True), {'C': [0.1, 1, 10]}),
    'DT': (DecisionTreeClassifier(), {'max_depth': [3, 5, 7]}),
    'MLP': (MLPClassifier(max_iter=1000), {'hidden_layer_sizes': [(50,), (100,)]})
}
```

```python
# Cargar datos preprocesados
X1_train, y1_train, X1_test, y1_test = joblib.load('../data/dataset1_processed.joblib')
X2_train, y2_train, X2_test, y2_test = joblib.load('../data/dataset2_processed.joblib')
X3_train, y3_train, X3_test, y3_test = joblib.load('../data/dataset3_processed.joblib')
```

```python
# Entrenar y guardar el mejor modelo para cada dataset y tipo
results = {}
for ds_name, (X_train, y_train, X_test, y_test) in zip(
    ['dataset1', 'dataset2', 'dataset3'],
    [(X1_train, y1_train, X1_test, y1_test), (X2_train, y2_train, X2_test, y2_test), (X3_train, y3_train, X3_test, y3_test)]
):
    print(f'Entrenando modelos para {ds_name}')
    best_score = 0
    best_model = None
    best_name = ''
    for name, (model, grid) in models_dict.items():
        print(f'  {name}...')
        if grid:
            m, params, score = models.tune_hyperparameters(model, grid, X_train, y_train)
        else:
            m = models.train_model(model, X_train, y_train)
            score = m.score(X_test, y_test)
            params = {}
        print(f'    Score: {score:.3f}')
        if score > best_score:
            best_score = score
            best_model = m
            best_name = name
        # Guardar cada modelo individual
        joblib.dump(m, f'../data/{ds_name}_{name}_model.joblib')
    print(f'Mejor modelo para {ds_name}: {best_name} (score={best_score:.3f})')
    joblib.dump(best_model, f'../data/{ds_name}_best_model.joblib')
    results[ds_name] = (best_name, best_score)

print('Entrenamiento y guardado de modelos individuales finalizado.')
```
