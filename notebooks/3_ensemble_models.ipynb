```python
# Cargar los mejores modelos individuales entrenados
import os

def load_best_models(ds_name, model_names):
    models_list = []
    for name in model_names:
        path = f'../data/{ds_name}_{name}_model.joblib'
        if os.path.exists(path):
            models_list.append((name, joblib.load(path)))
    return models_list

# Definir los modelos base para cada dataset (ajustar según resultados previos)
base_models1 = load_best_models('dataset1', ['LR', 'KNN', 'SVM', 'DT', 'MLP'])
base_models2 = load_best_models('dataset2', ['KNN', 'SVM', 'DT', 'MLP'])
base_models3 = load_best_models('dataset3', ['LR', 'NB', 'KNN', 'MLP'])
```

```python
# Cargar datos preprocesados
X1_train, y1_train, X1_test, y1_test = joblib.load('../data/dataset1_processed.joblib')
X2_train, y2_train, X2_test, y2_test = joblib.load('../data/dataset2_processed.joblib')
X3_train, y3_train, X3_test, y3_test = joblib.load('../data/dataset3_processed.joblib')
```

```python
# Crear y entrenar modelos ensemble (stacking y voting)
# Dataset 1
stack1 = ensemble.create_stacking_classifier(base_models1, LogisticRegression())
stack1.fit(X1_train, y1_train)
vote1 = ensemble.create_voting_classifier(base_models1)
vote1.fit(X1_train, y1_train)
joblib.dump(stack1, '../data/dataset1_stacking_model.joblib')
joblib.dump(vote1, '../data/dataset1_voting_model.joblib')

# Dataset 2
stack2 = ensemble.create_stacking_classifier(base_models2, LogisticRegression())
stack2.fit(X2_train, y2_train)
vote2 = ensemble.create_voting_classifier(base_models2)
vote2.fit(X2_train, y2_train)
joblib.dump(stack2, '../data/dataset2_stacking_model.joblib')
joblib.dump(vote2, '../data/dataset2_voting_model.joblib')

# Dataset 3
stack3 = ensemble.create_stacking_classifier(base_models3, LogisticRegression())
stack3.fit(X3_train, y3_train)
vote3 = ensemble.create_voting_classifier(base_models3)
vote3.fit(X3_train, y3_train)
joblib.dump(stack3, '../data/dataset3_stacking_model.joblib')
joblib.dump(vote3, '../data/dataset3_voting_model.joblib')

print('Modelos ensemble entrenados y guardados.')
```
