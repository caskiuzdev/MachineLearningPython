# Despliegue Web con Streamlit

Este notebook explica cómo ejecutar la aplicación web con Streamlit para predicción de riesgo de infarto usando los modelos entrenados.

## Instrucciones para ejecutar la app web

1. Asegúrate de haber ejecutado todos los notebooks anteriores y tener los modelos guardados en la carpeta `data/`.
2. Abre una terminal en la raíz del proyecto.
3. Ejecuta el siguiente comando:

```
streamlit run src/app.py
```

4. Se abrirá una interfaz web donde podrás seleccionar el modelo, ingresar los datos y obtener la predicción de riesgo de infarto.

Puedes personalizar la interfaz y los modelos cargados editando el archivo `src/app.py`.
